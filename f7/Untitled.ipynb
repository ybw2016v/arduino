```python
#!/usr/bin/env python3
import serial
import numpy as np
import time
ser = serial.Serial('/dev/ttyUSB0', 9600, timeout=1)

try:
    ser.open()
except:
    pass
```

```python
shu=np.zeros([180,180])
```

```python
def saomiao(nmp):
    ser.write('s'.encode())
    ser.flushInput()
    for i in range(0,180):
        time.sleep(1);
        
        shi = str(ser.readline())
        shi=shi[2:len(shi)-5]
        shi=shi.strip()
        shi=shi.split(',')
#         print(shi)
        nmp[i,:]=np.array(shi)
#         print(shi)
        
        pass
    return nmp
    pass
```

```python
saomiao(shu)
```
